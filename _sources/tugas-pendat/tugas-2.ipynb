```python
import numpy as np
import pandas as pd
from sklearn import datasets
```

```python
iris = pd.read_csv('https://raw.githubusercontent.com/Iqbalsuryaa/datamining/main/Iris.csv')
iris.head()
```

```python
iris.shape
```

```python
from sklearn.preprocessing import KBinsDiscretizer
a = KBinsDiscretizer(n_bins=4, encode='ordinal', strategy='uniform')
iris["SepalLengthCm"] = a.fit_transform(iris[["SepalLengthCm"]])
```

```python
iris["SepalLengthCm"]
```

```python
from sklearn.preprocessing import KBinsDiscretizer
a = KBinsDiscretizer(n_bins=4, encode='ordinal', strategy='uniform')
iris["SepalWidthCm"] = a.fit_transform(iris[["SepalWidthCm"]])
```

```python
iris["SepalWidthCm"]
```

```python
from sklearn.preprocessing import KBinsDiscretizer
a = KBinsDiscretizer(n_bins=4, encode='ordinal', strategy='uniform')
iris["PetalLengthCm"] = a.fit_transform(iris[["PetalLengthCm"]])
```

```python
iris["PetalLengthCm"]
```

```python
from sklearn.preprocessing import KBinsDiscretizer
a = KBinsDiscretizer(n_bins=4, encode='ordinal', strategy='uniform')
iris["PetalWidthCm"] = a.fit_transform(iris[["PetalWidthCm"]])
```
